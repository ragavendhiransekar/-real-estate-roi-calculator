```python
# 🏠 Real Estate ROI Calculator

def calculate_roi(price_lakhs, rent, expenses):
    price = price_lakhs * 100000  # convert lakhs to rupees
    annual_rent = rent * 12
    net_income = annual_rent - expenses
    roi = (net_income / price) * 100
    return round(roi, 2)

print("🏠 Real Estate ROI Calculator\n")

# Inputs
price = float(input("Enter Property Price (in Lakhs): "))
rent = float(input("Enter Monthly Rent: "))
expenses = float(input("Enter Annual Expenses: "))

# Calculation
roi = calculate_roi(price, rent, expenses)

# Decision Logic
if roi > 8:
    decision = "BUY ✅"
elif roi > 4:
    decision = "HOLD ⚖️"
else:
    decision = "AVOID ❌"

# Output
print("\n📊 RESULT")
print("ROI:", roi, "%")
print("Decision:", decision)
```
